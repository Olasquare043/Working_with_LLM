```python
from google import genai
from dotenv import load_dotenv
import os

load_dotenv()
```

```python
# chat function
def chat(prompt):
    # initializing
    client = genai.Client(api_key=os.getenv('gemini_api'))
    response = client.models.generate_content(
        model="gemini-2.5-flash" ,contents=prompt
    )
    print(response.text)
```

```python
while True:
    prompt=input("Welcome to Olasquare chatbot, How may I help you today?\n press X to exit")
    if prompt == "X".lower():
        exit
    print(chat(prompt))
```
